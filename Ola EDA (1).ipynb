```python
revenue=df['Booking Value'].sum()
print("Total Revenue:",round(revenue,0))
```

```python
revenue_by_vehicles=df.groupby('vehicle_type')['Booking Value'].sum().sort_values(ascending=False)
print("Revenue By Vehicles:",round(revenue_by_vehicles,0))
```

```python
plt.figure(figsize=(15,6))
colors=('yellow','blue','brown','pink','orange','red','green')
sb.barplot(x=revenue_by_vehicles.index,y=revenue_by_vehicles.values,palette=colors)
plt.title("Revenue By Vehicles Type")
plt.ylabel("Total Revenue")
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```
